#### Words Used Once

```python
# Frequency distribution of words in corpus
fdist = FreqDist(tokens)
```

```python
# List of words used only once in corpus
used_once = fdist.hapaxes()
len(used_once)
```

```python
# Add new stop words
from sklearn.feature_extraction import text 

stop_words = text.ENGLISH_STOP_WORDS.union(domain_stopwords, used_once)
```

```python
# This stop words list should cut down on original 30,526-word vocabulary!
len(stop_words)
```

```python
# Find most commonly-used RELEVANT words (those not in domain_stopwords)
#most_used = [item[0] for item in fdist.most_common(35) if not item[0] in domain_stopwords]
# Most of the words look like they're describing Cabernet Sauvignon, which makes sense, 
# since that is the most commonly-reviewed varietal in the dataset
```

##### Filter descriptions down to about 500 most-used words to decrease dimensions (after removing stopwords)

```python
most_used_500 = [item[0] for item in fdist.most_common(625) if not item[0] in stop_words]
```

```python
len(most_used_500)
```

```python
def filterDescription(text):
    text = [word for word in text.split() if word in most_used_500]
    return ' '.join(text)
```

```python
df.description = df.description.apply(filterDescription)
```

## Count Vectorizer II

```python
# Create a document-term matrix using CountVectorizer
cv = CountVectorizer(stop_words=stop_words)
data_cv = cv.fit_transform(df.description)

data_dtm = pd.DataFrame(data_cv.toarray(), columns=cv.get_feature_names())
data_dtm.index = df.index


print(f"Dimensions of Document-term matrix: {data_cv.toarray().shape}")
data_dtm.head()
```

## TF-IDF

```python
tfidfvec = TfidfVectorizer(stop_words=stop_words)#ngram_range=(1,2))
X_tfidf = tfidfvec.fit_transform(df.description)

tfidf_dtm = pd.DataFrame(X_tfidf.toarray(), columns=tfidfvec.get_feature_names())
tfidf_dtm.index = df.index

print(f"Dimensions of Document-term matrix: {X_tfidf.toarray().shape}")
tfidf_dtm.head()
```

## Dimension Reduction

#### LSA by SVD

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.random_projection import sparse_random_matrix

svd = TruncatedSVD(n_components=30, n_iter=15, random_state=42)
svd.fit(X_tfidf) 
svdfeatures_train = svd.transform(X_tfidf)
```

```python
plt.plot(svd.explained_variance_ratio_)
plt.xlabel('# components')
plt.ylabel('explained variance');
plt.title('Scree Plot for Wine Description Data');
```

```python
# Looks like somewhere between 5-10 components
```

```python
plt.plot(np.cumsum(svd.explained_variance_ratio_))
plt.xlabel('# components')
plt.ylabel('cumulative explained variance');
plt.title('Cumulative Explained Variance by SVD');
```

```python
svd1 = TruncatedSVD(10) 
doc_topic = svd1.fit_transform(X_tfidf)
svd1.explained_variance_ratio_
```

```python
topic_word = pd.DataFrame(svd1.components_.round(3),
             index = ["component_01","component_02", "component_03", 
                      "component_04", "component_05","component_06", 
                      "component_07", "component_08","component_09", 
                      "component_10"],
             columns = tfidfvec.get_feature_names())
topic_word
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

```python
display_topics(svd1, tfidfvec.get_feature_names(), 5)
```

```python
# gensim
from gensim import corpora, models, similarities, matutils
```

```python
# Convert sparse TF-IDF matrix to a gensim corpus
tX_tfidf = X_tfidf.transpose()
corpus = matutils.Sparse2Corpus(tX_tfidf)
```

```python
id2word = dict((v, k) for k, v in tfidfvec.vocabulary_.items())
```

```python
len(id2word)
```

```python
# Create lda model (equivalent to "fit" in sklearn)
lda = models.LdaModel(corpus=corpus, num_topics=5, id2word=id2word, passes=5)
```

```python
lda.print_topics()
```

```python
import pyLDAvis.gensim
from gensim.corpora import Dictionary
```

```python
gensim_dict = Dictionary([[word for word in desc.split() if not word in stop_words
                          ] for desc in list(df.description)])
```

```python
wine_vis_data = pyLDAvis.gensim.prepare(lda, corpus, gensim_dict)
```
